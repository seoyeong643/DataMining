```python
import pandas as pd
import numpy as np
df = pd.read_csv("24_train_3.csv")
df
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
import json
import re
from collections import Counter
```

```python
#sample code to generate unigrams for each document
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import *

# stemming tool from nltk
stemmer = PorterStemmer()
# a mapping dictionary that help remove punctuations
remove_punctuation_map = dict((ord(char), None) for char in string.punctuation)
def get_tokens(text):
  # turn document into lowercase
  lowers = text.lower()
  # remove punctuations
  no_punctuation = lowers.translate(remove_punctuation_map)
  # tokenize document
  tokens = nltk.word_tokenize(no_punctuation)
  # remove stop words
  filtered = [w for w in tokens if not w in stopwords.words('english')]
  # stemming process
  stemmed = []
  for item in filtered:
      stemmed.append(stemmer.stem(item))
  # final unigrams
  return stemmed
```

```python
#Process all documents/articles
all_tokens = [get_tokens(text) for text in df['Text']]
print(all_tokens)
```

```python
#load dictionary
with open("dictionary.txt", 'r') as f:
    dictionary = [word.strip() for word in f]
print(dictionary)
```

```python
# Term frequency, TF_ij is defined as where the term frequency of word j is normalized by the max num of occurences of any word, k, in the same doc
def compute_tf(tokens, dictionary):
    # only consider words in dictionary.txt
    valid_tokens = [token for token in tokens if token in dictionary]

    word_counts = Counter(valid_tokens)
    max_freq = max(word_counts.values()) if word_counts else 1

    tf_dict = {word: round(count/max_freq, 4) for word, count in word_counts.items()}
    return tf_dict

tf_dicts = [compute_tf(tokens, dictionary) for tokens in all_tokens]
print(tf_dicts)
```

```python
#To compute inverse doc frequency, let m_j be the num of docs in which word j appears
doc_count = len(all_tokens)
word_doc_count = {}

for tokens in all_tokens:
    #unique words in document that are in dictionary
    unique_words = set(token for token in tokens if token in dictionary)
    for word in unique_words:
        word_doc_count[word] = word_doc_count.get(word,0)+1

idf_dict = {word: round(np.log10(doc_count/count), 4)
            for word, count in word_doc_count.items()}
print(idf_dict)
```

```python
num_docs = len(tf_dicts)
num_terms = len(dictionary)

tfidf_matrix = np.zeros((num_docs, num_terms))

for i, tf_dict in enumerate(tf_dicts):
    for j, term in enumerate(dictionary):
        if term in tf_dict and term in idf_dict:
            tfidf_matrix[i,j] = round(tf_dict[term] * idf_dict[term], 4)

np.savetxt('matrix.txt', tfidf_matrix, delimiter=',', fmt='%.4f')
```

```python
#Compute top 3 frequency and TFIDF words in each category
def top_words(df, all_tokens, dictionary, tfidf_matrix=None):
    categories = df['Category'].unique()
    frequency_results = {}
    tfidf_results = {}
    dictionary_list = sorted(dictionary)

    for category in categories:
        category_indices = df[df['Category'] == category].index

        # Get frequency counts
        word_counts = Counter()
        for idx in category_indices:
            word_counts.update(token for token in all_tokens[idx] if token in dictionary)

        # Top 3 frequency
        top_freq = {word: int(count) for word, count in
                   sorted(word_counts.items(), key=lambda x: x[1],
                         reverse=True)[:3]}
        frequency_results[category] = top_freq

        # Top 3 avg TFIDF
        if tfidf_matrix is not None:
            category_tfidf = tfidf_matrix[category_indices]
            avg_tfidf = np.mean(category_tfidf, axis=0)
            top_indices = np.argsort(avg_tfidf)[-3:][::-1]

            top_tfidf = {dictionary_list[i]: round(float(avg_tfidf[i]), 4)
                        for i in top_indices}
            tfidf_results[category] = top_tfidf

    return frequency_results, tfidf_results
```

```python
freq_results, tfidf_results = top_words(df, all_tokens, dictionary, tfidf_matrix)
```

```python
with open('frequency.json', 'w') as f:
    json.dump(freq_results, f, indent=2)

with open('scores.json', 'w') as f:
    json.dump(tfidf_results, f, indent=2)
```
